```python
data_new = data.drop(cat_features, axis=1)
```

```python
data_new['status']=data_new['status'].apply(lambda x: 0 if '< 0' in x else  1 if '< 200' in x else 2)
data_new['savings']=data_new['savings'].apply(lambda x: 0 if 'unknown' in x else  1 if '<  100' in x else 2 if '<  500' in x else 3 if '< 1000' in x else 4)
data_new['employment_duration']=data_new['employment_duration'].apply(lambda x: 0 if '< 1' in x else  1 if '< 4' in x else 2 if '< 7' in x else 3)
data_new['number_credits']=data_new['number_credits'].apply(lambda x: 0 if '>= 6' in x else  1 if '4-5' in x else 2 if '2-3' in x else 3)
```

```python
data_new=pd.get_dummies(data_new, columns=bin_features, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
def euclidean_distance(X, y):
    return np.sqrt(np.sum((X - y) ** 2, axis=1))

def predict(X_train, y_train, X_test, k):
    ans = list()
    for vec in X_test:
        ans.append((euclidean_distance(X_train, vec)).argsort()[:k])
    
    prediction = list()
    for i in ans:
        prediction.append(np.bincount(y_train[i]).argmax())
        
    return np.array(prediction).reshape(-1, 1)
    
def loss(X_train,y_train, p):
    p = np.clip(p, 1e-10, 1 - 1e-10)
    return np.mean(y * np.log(p) + (1 - y) * np.log(1 - p))
```

```python
X = data_new.drop('credit_risk_good', 1).values
y = data_new['credit_risk_good'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
scaler = StandardScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.fit_transform(X_test)
```

```python
for i in range(3,10):
  y_pred=predict(X_train_scale,y_train,X_test_scale,i)
  print(accuracy_score(y_test, y_pred))
```
