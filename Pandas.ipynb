```python
data_info = pd.DataFrame()
data_marks = pd.DataFrame()
for x in os.listdir('/'):
    if 'info' in x:
        data_info = pd.concat([data_info, pd.read_csv('/' + x)])
    if 'marks' in x:
        data_marks = pd.concat([data_marks, pd.read_csv('/' + x)])
data = pd.merge(data_info, data_marks, on='index')

data.head(6)
```

```python
%%time
count = dict()
sigma = dict()

for x in data[['race/ethnicity', 'reading score']].to_numpy():
    if x[0] not in sigma:
      sigma[x[0]] = 0
      count[x[0]] = 0
    else:
      sigma[x[0]] += x[1]
      count[x[0]] += 1

for key in sigma:
    print(key, sigma[key] / count[key])

    # в 2 раза больше времени
```

```python
%%time
minimum = dict()

for x in data[['education', 'writing score']].to_numpy():
    if x[0] not in minimum:
      minimum[x[0]] = 100000
    else:
      minimum[x[0]] = min(minimum[x[0]],x[1])

for key in minimum:
    print(key, minimum[key])

    # в 2 раза больше времени
```

```python
data["Writing_PassStatus"]=data['writing score'].apply(lambda x: 'P' if x>=passmark else 'F')
data["Writing_PassStatus"].value_counts()
```

```python
data["Reading_PassStatus"]=data['reading score'].apply(lambda x: 'P' if x>=passmark else 'F')
data["Reading_PassStatus"].value_counts()
```

```python
data["OverAll_PassStatus"] = (data[['Math_PassStatus', 'Writing_PassStatus', 'Reading_PassStatus']] == 'P').apply(lambda x: all(x), axis=1)
data["OverAll_PassStatus"]=data['OverAll_PassStatus'].apply(lambda x: 'P' if x==True else 'F')
```

```python
data["OverAll_PassStatus"].value_counts()
```

```python
def GetGrade(average_mark):
    mark = 'F'
    if average_mark >= 90:
        mark = 'A'
    elif average_mark >= 80:
        mark = 'B'
    elif average_mark >= 70:
        mark = 'C'
    elif average_mark >= 60:
        mark = 'D'
    elif average_mark >= 50:
        mark = 'E'      
    return mark   

data['Grade'] = data[['math score', 'reading score', 'writing score']].mean(axis=1).apply(GetGrade)
```
